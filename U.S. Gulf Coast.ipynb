```python
import requests
from bs4 import BeautifulSoup

url = "https://www.eia.gov/dnav/pet/hist/EER_EPD2DXL0_PF4_RGC_DPGD.htm"
response = requests.get(url)
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
links = soup.find_all()

# Extracting text content from the links
for link in links:
    print(link.get('href')) 
    print(link.text.strip()) 
```

```python
import psycopg2

# Connect to PostgreSQL database
conn = psycopg2.connect(
    dbname="US",
    user="postgres",
    password="123456",
    host="localhost",
    port="5432"
)
cur = conn.cursor()
data = [
    ('https://www.eia.gov/dnav/pet/hist/EER_EPD2DXL0_PF4_RGC_DPGD.htm', 'Fuel'),
]
```

```python
try:
    create_table_sql = '''
        CREATE TABLE Crude_Oil_Prices (
            id SERIAL PRIMARY KEY,
            url VARCHAR(255),
            category VARCHAR(50)
        )
    '''
    cur.execute(create_table_sql)  # Creating the table

    sql = "INSERT INTO Crude_Oil_Prices (url, category) VALUES (%s, %s)"

    for row in data:
        cur.execute(sql, row)

    conn.commit()  

except (Exception, psycopg2.DatabaseError) as error:
    conn.rollback()  
    print(error)
```

```python
try:
    cur.execute("SELECT * FROM Crude_Oil_Prices")
    rows = cur.fetchall()

    for row in rows:
        print(row)  

except (Exception, psycopg2.DatabaseError) as error:
    conn.rollback() 
    print(error)
```
